```python
# Run a simple linear regression 
# Independent variables x
X = df[['Demo','Demo1-3','Demo4-5']]
# Dependent variable y
y = df['Units_Sold']
```

```python
import statsmodels.api as sm

# Statsmodels don't automatically fit a constant, so we need to use the method sm.add_constant() to add a constant column 1
X = sm.add_constant(X)
X
```

```python
# Adding Y fit constant
y = sm.add_constant(y)
y
```

```python
# Creating a fitted model with Demo
X = df[['Average_Retail_Price','Sales_Rep','Demo','Demo1-3','Demo4-5']]
# Dependent variable y
y = df['Units_Sold']
X = sm.add_constant(X)


model_1 = sm.OLS(y,X).fit()
intercept = model_1.params[0]
print(model_1.summary())


# P value for the demos is 0.000, which is *** significant. Demos are statistically significant.
```

```python
# Run a simple linear regression 
# Independent variables x
X = df['Endcap']
# Dependent variable y
y = df['Units_Sold']
```

```python
import statsmodels.api as sm

# Statsmodels don't automatically fit a constant, so we need to use the method sm.add_constant() to add a constant column 1
X = sm.add_constant(X)
X
```

```python
# Creating a fitted model with EndCap promotion

model_2 = sm.OLS(y,X).fit()
print(model_2.summary())

#y = 240.6958 + 343.2292  * Endcap
# P value for Endcap is 0.000, which is *** significant. Endcap promotion is statistically significant.
```
